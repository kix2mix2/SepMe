```python
from sklearn.datasets import make_swiss_roll
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pydiffmap import diffusion_map as dm
from pydiffmap.visualization import embedding_plot, data_plot
%matplotlib inline
```

```python
df = pd.read_csv('swiss_roll.csv')
df.head()
```

```python

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df.iloc[:,0], df.iloc[:,1], df.iloc[:,2], cmap='viridis', c=df['color'])


ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
fig = plt.figure(figsize=(30,10))
ax = fig.add_subplot(131)
ax.scatter(df.iloc[:,0], df.iloc[:,1], c = df['color'])

ax = fig.add_subplot(132)
ax.scatter(df.iloc[:,0], df.iloc[:,2], c = df['color'])

ax = fig.add_subplot(133)
ax.scatter(df.iloc[:,1], df.iloc[:,2], c = df['color'])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
df_PCA = pca.fit_transform(df.loc[:,['x','y','z']])


fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.scatter(df_PCA[:,0], df_PCA[:,1], c = df.color)
```

```python

from sklearn.manifold import TSNE

X_embedded = TSNE(n_components=2).fit_transform(df.iloc[:,0:3])
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.scatter(X_embedded[:,0], X_embedded[:,1], c = df.color)
```

```python
from sklearn.manifold import Isomap

X_embedded = Isomap(n_components=2).fit_transform(df.iloc[:,0:3])
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.scatter(X_embedded[:,0], X_embedded[:,1], c = df.color)
```
